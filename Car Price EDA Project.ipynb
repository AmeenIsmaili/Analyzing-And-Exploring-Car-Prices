# Q6. What are the primary central tendency methods are used for filling the categorical and numerical missing values?

```python
# Several methods are used to fill missing values, however, central tendency(mean, median, mode) for numeric datasets
# and mode imputation or forward and backward filling for catagorical datasets are most commonly used methods to 
# fill missing values.
```

# Q7. Fill the missing values of “drivewheel” and “wheelbase” using the appropriate central tendency method with memory updating.

```python
df3[["drivewheel", "wheelbase"]].isna().sum()
```

```python
df3["wheelbase"].fillna(df3["wheelbase"].mean(), axis=0, inplace=True)
df3["drivewheel"].fillna(df3["drivewheel"].mode()[0], axis=0, inplace=True )
df3[["drivewheel", "wheelbase"]].isna().sum()
```

# Q21. How does 'enginesize' impact 'price,' and is it a cost-effective feature for customers?

```python
df6_eng_price = df6[["enginesize", "price"]]
df6_eng_price.groupby('enginesize')['price'].value_counts()
```

```python
Figure = plt.figure(figsize=(20, 8))

sns.boxplot(x=df6_eng_price['enginesize'], y=df6_eng_price['price'])
plt.title("Engine Size vs Car Price")
plt.show()
```

```python
Figure = plt.figure(figsize=(6, 3))

sns.heatmap(df6_eng_price.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
Figure = plt.figure(figsize=(6, 3))

sns.scatterplot(x = df6_eng_price["enginesize"], y = df6_eng_price["price"])
plt.grid()
plt.show()
```

```python
# Observation:  As the visuals suggests, there is strong positive relationship between 'enginesize' and 'price'.
# The bigger the enginesize the higher the price and vice versa. 
```
